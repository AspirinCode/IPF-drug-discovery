```python
import json
from requests import get
```

```python
query_json = {
    "command": "correlation",
    "parameters":{
        "experimentdesign": "195785",
        "q": "idiopathic pulmonary fibrosis",
        "apikey": "6697b509829731cd9537e953648c5a7d",
        "organism": "1",
        "datatype": "1",
        "v": "0",
        "samplesource": [
          "195775",
          "195776",
          "195777",
          "195780"]
        }
    }
query_json['parameters']['q'] = 'bipolar disorder'
```

```python
def json2xml(query_json):
    base_url = "https://jeggalab.ussc.informatics.illumina.com/c/nbapi/"
    xml = base_url + query_json["command"] + ".api?"
    xml += "apikey=" + query_json["parameters"]["apikey"]
    for key in query_json["parameters"].keys():
        if isinstance(query_json["parameters"][key],list):
            for i in query_json["parameters"][key]:
                xml += "&" + key+ '=' + i
        else:
            if key == 'apikey':
                continue
            elif key == 'q':
                string_to_add = query_json["parameters"][key].replace(' ','+')
            else:
                string_to_add = query_json["parameters"][key]
            xml += "&" + key+ '=' +string_to_add
    return xml
```

```python
xml = json2xml(query_json)
ce_result = json.loads(get(xml).text)
ce_result['result']
```
